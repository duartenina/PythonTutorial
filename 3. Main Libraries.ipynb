# NumPy

NumPy is a library focused on fast array math. Most other big python libraries accept or require numpy arrays.

```python
import numpy as np
```

```python
x = np.array([1, 2, 3])
display(x)
display(x*2 + 1)
```

```python
x = np.linspace(-1.5 * np.pi, 1.5 * np.pi, 100)

plt.plot(x, np.sin(x), label='sin')
plt.plot(x, np.cos(x), label='cos')
plt.plot(x, np.tanh(x), label='tanh')

plt.legend()
```

```python
mat = np.zeros((3, 4))
display(mat)
for row in range(mat.shape[0]):
    for col in range(mat.shape[1]):
        mat[row, col] = col * 10 + row
display(mat)
mat[:, 0] += 100
display(mat)
mat[1, :] /= 2
display(mat)
```
