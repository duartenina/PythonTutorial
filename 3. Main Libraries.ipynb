# NumPy

NumPy is a library focused on fast array math. Most other big python libraries accept or require numpy arrays (`ndarray`).

```python
import numpy as np
```

#### Generate Arrays

```python
display(np.array([1, 2, 3]))
display(np.arange(20, 2, -3))
display(np.linspace(0, np.pi, 5))
display(np.linspace(0, 1, 10, endpoint=False))
display(np.zeros(10))
display(np.ones(10))
display(np.random.rand(5))
```

```python
a = np.linspace(11, 15, 5)
display(a)
display(a[2])
display(a[2:4])
display(a[[0, 1, 4]])
display(a[[-1, -3]])
display(a[[False, True, True, False, True]])
```

```python
a = np.arange(1, 11)
cond = a < 4
display(a[cond])
display(cond)
display(np.nonzero(cond))
display(a[~cond])
display(a[(a > 5) & (a % 2 == 0)])
```

```python
a = np.random.randint(0, 2, 6, dtype=bool)
display(a)
print(a.sum())
print(a.any())
print(a.all())
```

#### N-d Arrays

```python
array_3d = np.random.random((2,3,4))
display(array_3d)
print('ndim  =', array_3d.ndim)  # Number of dimensions
print('shape =', array_3d.shape) # tuple with size of all dimensions
print('len   =', len(array_3d))  # length of first dimension
print('size  =', array_3d.size)  # number of elements
```

```python
a = np.array([1, 2, 3, 4])
display(a)
display(a.reshape((2, 2)))
display(np.broadcast_to(a, (2, 4)))
display(np.tile(a, 2))              # tile the whole array
display(np.repeat(a, 2, axis=0))    # repeat each individual element

b = np.array([[1, 2], [3, 4], [5, 6]])
display(b)
display(b.T)
```

#### Types

`ndarrays` can be of many possible types ([np.dtype](https://docs.scipy.org/doc/numpy/reference/arrays.dtypes.html)).

```python
for t in [int, float, complex, str, bool, 'unicode']:
    print('{:20}: {}'.format(str(t), np.zeros(5, dtype=t)))
```

And even struture-like arrays:

```python
first_names = ['John', 'Terry', 'Michael', 'Graham', 'Eric', 'Terry']
last_names = ['Cleese', 'Gilliam', 'Palin', 'Chapman', 'Idle', 'Jones']
years_of_birth = [1939, 1940, 1943, 1941, 1943, 1942]
years_of_death = [0, 0, 0, 1989, 0, 0]
table = [(first, last, birth, death) for first, last, birth, death in zip(first_names, last_names, years_of_birth, years_of_death)]
print(table)
np_table = np.array(table, dtype=[('first_name', 'U32'), ('last_name', 'U32'), ('year_of_birth', int), ('year_of_death', int)])
display(np_table)
print(np_table['first_name'])
print(np_table['year_of_birth'])
print('Ordered by Year of Birth:')
for info in np.sort(np_table, order='year_of_birth'):
    print(' *', info)
print(np_table[2])
```

## Math with `ndarrays`

```python
x = np.linspace(-1.5 * np.pi, 1.5 * np.pi, 100)

plt.plot(x, x/5, label='x/5')
plt.plot(x, np.sin(x), label='sin')
plt.plot(x, np.tanh(x), label='tanh')
plt.plot(x, np.exp(x - 3) - 0.75, label='exp', color='C3')
plt.axhline(-0.75, ls='--', color='C3')
plt.axhline(1-0.75, ls='--', color='C3')
plt.axvline(3, ls='--', color='C3')

plt.legend(loc='upper left')
plt.ylim(-1.1, 1.1)
```

```python
mat = np.zeros((3, 4))
display(mat)
for row in range(mat.shape[0]):
    for col in range(mat.shape[1]):
        mat[row, col] = col * 10 + row
display(mat)
display(mat[:, 1])
display(mat[2, :])
display(mat[:, :2])
display(mat[1:, :])
display(mat[1:, :2])
mat[1, :] /= 2
display(mat)
```

```python
a = np.arange(5)
b = a
c = a.copy()
a[3] = -1
display(a)
display(b)
display(c)
```

## More Complicated Examples

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-7.5, 7.5, 100)
y = np.linspace(-7.5, 7.5, 50)
xx, yy = np.meshgrid(x, y)
print(x.shape, y.shape, xx.shape, yy.shape)
zz = np.sin(np.sqrt(xx**2 + yy**2))

fig = plt.figure(figsize=(11, 5))

ax1 = plt.subplot(121, projection='3d')

ax1.contour(xx, yy, zz, 50)

ax2 = plt.subplot(122)

plt.pcolormesh(xx, yy, zz)
plt.colorbar()
plt.axis('equal')

plt.show()

fig = plt.figure(figsize=(9, 8))

ax = plt.subplot(111, projection='3d')

c = ax.plot_surface(xx, yy, zz, cmap='plasma')
c = ax.contourf(xx, yy, zz, 100, cmap='plasma', zdir='z', offset=-1)
plt.colorbar(c)

plt.show()
```

```python
table = np.genfromtxt('shots_settings.csv', comments='#', delimiter=',', dtype=('U16,u4,f,f,f,f'), skip_header=1)
table = np.sort(table, order='f0')
display(table)

for col in ['f2', 'f3']:
    print('{}: {:.2f} \u00b1 {:.2f}'.format(col, np.mean(table[col]), np.std(table[col])))
    
x, y = table['f1'], table['f2']
p = np.polyfit(x, y, 1)

fig = plt.figure(figsize=(6, 4))
plt.plot(x, y, 'o')
plt.plot(x, np.polyval(p, x))
plt.text(0.55, 0.4, 'm = {:.2e}\nb = {:.2f}'.format(*p), ha='left', va='top', transform=fig.transFigure, color='C1', fontdict={'size': 16})

plt.show()
```
